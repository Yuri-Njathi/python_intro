```python
blurred_image = cv2.GaussianBlur(img, (5, 5), 0)
#plt.tick_params(left = False, right = False , labelleft = False , labelbottom = False, bottom = False)
plt.imshow(blurred_image,cmap="gray")
plt.show()
```

# Edge Detection

Highlight the edges in an image using Canny edge detector

```python
img_canny = cv2.imread('/content/Basics_of_Images_Samples/Pixel-Art.png', cv2.IMREAD_GRAYSCALE)
edges = cv2.Canny(img_canny, threshold1=100, threshold2=200)#, cv2.COLOR_RGB2GRAY)
#plt.tick_params(left = False, right = False , labelleft = False , labelbottom = False, bottom = False)
plt.imshow(edges,cmap="gray")
plt.show()
```

# Thresholding

Converts a grayscale image into a binary image

Pixels above a certain threshold are set to 255 and below to 0.

```python
_, binary_image = cv2.threshold(gray_image*255,64,255,cv2.THRESH_BINARY) #multiplied by 255 bcoz image is float32.
#plt.tick_params(left = False, right = False , labelleft = False , labelbottom = False, bottom = False)
plt.imshow(binary_image,cmap="gray")
plt.show()
```
